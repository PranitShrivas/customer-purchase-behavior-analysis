```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior (1).csv')

df.head()
```

```python
## To check overall structure of the data.
df.info()
```

```python
## to check the summary of this data.(count,min,max,percentile and deviation)
df.describe()
```

```python
## to check summary colums for all(string and numbers)

df.describe(include='all')
```

```python
# now to check if there any missing values init..

df.isnull().sum()
```

```python
##we got null values in review rating column

df['Review Rating'] = df['Review Rating'].fillna(df['Review Rating'].median())
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Gender')['Review Rating'] \
.transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# lets do some formatting in column names from (Customer ID to customer_id)
df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
#rename 
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
```

```python
# lets create new column "age_group"

labels = ['young adults', 'adults', 'middle aged', 'seniors']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df.head()
```

```python
df[['age', 'age_group']].head()
```

```python
## new column - purchase_frequency_days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
